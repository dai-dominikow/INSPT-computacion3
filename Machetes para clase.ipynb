Derivadas

```python
def deriv(x,t):

  j= len(x) - 1
  i=0
  dx=np.zeros(len(x))
  
  while i < j:
    dx[i]= (x[i+1]-x[i])/(t[i+1]-t[i])
    i=i+1
  dx[i]=None

  return dx
  
v = deriv(x,t)
plt.plot(t, v, color= 'green')
plt.xlabel('tiempo')
plt.ylabel('velocidad')
plt.show()
```

# Simulaci√≥n din√°mica

Los sistemas din√°micos o el estudio din√°mico de los sistemas es el que refiere a sistemas que evolucionan en el tiempo.
Retrato de fase: ejemplo velocidad vs posici√≥n, la derivada de una variable respecto a dicha variable. Representacion geometrica de las trayectorias de un sistema dinamico en el plano.
Los retratos de fases revelan informacion sobre la existencia de atractores y repulsores.
Los sistemas din√°micos no lineales exhiben atractores, que se definen como puntos en el retrato de fases donde todas las trayectorias cercanas convergen, donde tiende a evolucionar el sistema.

### metodo de Euler:
Metodo de Euler:  CONSISTE EN SEGUIR LAS TANGENTES DE LA FUNCION EN CADA PUNTO se utiliza cuando tenemos la derivada de la funcion que representa la solucion de una EDO, es decir, su pendiente en cualquier punto. Este m√©todo es un m√©todo de primer orden, lo que quiere decir que cada variable de estado se calcular√° a partir de su valor en el tiempo anterior.
Un lector entrenado en matem√°tica, notar√° que estamos calculando una magnitud a partir de utilizar la derivada de la magnitud (v=dy/dt) con respecto al tiempo, multiplicada por el incremento del tiempo. Es decir, uniendo el punto conocido con el punto a calcular a partir de
una recta que tiene como pendiente la derivada de la posici√≥n en ese punto.

Metodo de integracion numerica para resolver ecuaciones diferenciales ordinarias a partir de un valor inicial dado.
Cada instante calculo la derivada de la posici√≤n con la derivada del punto anterior.
Sirve solamente para ecuaciones diferenciales de grado uno.
posicion --> velocidad --> aceleracion
Es un metodo de primer orden, lo que significa que el error local (diferencia que se produce entre el valor real de la funci√≥n y el aproximado mediante la recta tangente, en lugar de moverse por la curva, suponiendo que el punto donde se cruzan la curva real y la recta que la aproxima no tiene error alguno) es proporcional al cuadrado del paso y el erorr global es proporcional al tama√±o del paso (en general el error global tiene un orden menos que el local, es un poco mas grueso ya que los errores locales se acumulan dand lugar al global).

### metodo de Cromer:
La diferencia con el m√©todo de Euler est√°ndar es que el m√©todo de Euler semi-impl√≠cito usa v_n+1 en la ecuaci√≥n para x_n+1 , mientras que el m√©todo de Euler usa v_n.
Es mas estable en problemas de orbitas y oscilatorios, pero en cuanto a precision es igual de pobre.
Se calcula en otro orden
aceleracion --> velocidad --> posicion


```python
# por ejemplo, final de FT1

#cada instante calculo la derivada de la posici√≤n con la derivada del punto anterior

from vpython import *

G=6.67e-11      #constante universal
rA= 1.1903e6     #radio Sirius A
rB= 5844      #radio Sirius B
mA=3.978e30     #masa Sirius A
mB=2.025e30     #masa Sirius B
mu= mA*mB/(mA+mB)   #masa reducida
d= 3e9       #distancia entre las estrellas

#posici√≥n centros de masa
xA = (mB/(mA+mB))*d
xB = -(mA/(mA+mB))*d

#primero creo las esferas y les doy sus atributos
SiriusA = sphere(pos = vector(xA,0,0), 
                radius=rA*500,    #multiplique 10 veces el radio para que se vea mejor
                color=color.cyan, 
                make_trail=True)
                
SiriusB = sphere(pos=vector(xB,0,0),
              radius=rB*500,
              color=color.red,
              make_trail=True
              )

vB = sqrt(G*mA*mag(SiriusB.pos)/(d**2))

#momento inicial del sistema
SiriusA.p = vector(0,0,0) 
SiriusB.p = mB*vector(0,vB,0)  #p=m*v

#defino mi tiempo
t = 0
dt = 100
t_f=24*3600*30

#momento angular
l=mag(cross(SiriusA.pos,SiriusA.p)+cross(SiriusB.pos,SiriusB.p))

# Metodo de Cromer - Euler
while t<t_f:
    rate(50) #velocidad de la simulacion
    r = SiriusB.pos - SiriusA.pos
    F = -G*mA*mB*norm(r)/mag(r)**2 
    SiriusB.p = SiriusB.p + F*dt
    SiriusA.p = SiriusA.p - F*dt
    SiriusB.pos = SiriusB.pos + SiriusB.p*dt/mB
    SiriusA.pos = SiriusA.pos + SiriusA.p*dt/mA
    t = t + dt
```

### FFT
La transformada r√°pida de Fourier, conocida por la abreviatura FFT (del ingl√©s Fast Fourier Transform) es un algoritmo eficiente que permite calcular la transformada de Fourier discreta (DFT) y su inversa. La FFT es de gran importancia en una amplia variedad de aplicaciones, desde el tratamiento digital de se√±ales y filtrado digital en general a la resoluci√≥n de ecuaciones en derivadas parciales o los algoritmos de multiplicaci√≥n r√°pida de grandes enteros. 
Esta transformada convierte la se√±al del dominio del tiempo al dominio de la frecuencia,  devolviendo el espectro (Es decir, arm√≥nicos de qu√© frecuencia componen la se√±al).
la transformada de Fourier contiene todas las frecuencias del tiempo durante el cual existi√≥ la se√±al; es decir, en la transformada de Fourier se obtiene un solo espectro de frecuencias para toda la funci√≥n.
La antitransformada de Fourier permite, teniendo el espectro, recuperar la funci√≥n original. Esto es muy √∫til porque permite realizar operaciones sobre el espectro que ser√≠an m√°s dif√≠ciles de realizar en el dominio del tiempo. En el dominio de la frecuencia se pueden aplicar filtros para recortar o ‚Äúpodar‚Äù (es el t√©rmino t√©cnico) toda una frecuencia del sonido. El resultado de la transformada de Fourier es una funci√≥n compleja. 

Dada una se√±al periodica, podemos usar la transformada rapida de fourier para ver su contenido de frecuencias.
Graficamente, pasamos de una funcion dependiente del tiempo a una dependiente de las frecuencias en el plano complejo. El plot que hacemos solo refleja la parte real del complejo aunque tambien se puede laburar con la imaginaria, pero vamos a usar RFFT (que devuelve la parte real)
La transformada entonces reemplaza integrales por sumatorias (similar al metodo de los trapecios)

Un muestreo puede ser representado como una multiplicacion de la se√±al original (continua) con un pulso unitario.

¬øQu√© quiere decir muestreada? que la se√±al real por ser anal√≥gica tendr√≠a infinitos valores, pero nosotros no podemos representar infinitos valores en la pc, por lo que la frecuencia de muestreo establece cu√°ntos valores de tensi√≥n, intensidad o lo que sea que represente la se√±al se tomar√°n por segundo.

La frecuencia de muestreo debe cumplir con el teorema del Muestreo, o teorema de Nyquist-Shannon que explica la relaci√≥n entre la velocidad de muestreo y la frecuencia de la se√±al medida.
B√°sicamente dice que para reconstruir una se√±al, el muestreo debe tener el doble de frecuencia que la m√°xima de la se√±al.

Si una se√±al es muestreada a una velocidad de muestreo menor que el doble de la frecuencia Nyquist, los componentes de frecuencias falsas m√°s bajas aparecen en los datos muestreados. Este fen√≥meno se conoce como aliasing. Se produce cuando una velocidad de muestreo es demasiado baja y reproduce una representaci√≥n incorrecta de forma de onda.

Este teorema es muy espec√≠fico y habla de que si una se√±al peri√≥dica anal√≥gica tiene una banda de frecuencias definida, entonces muestreando con una frecuencia que es el doble de la frecuencia m√°xima de la se√±al se obtendr√° una cantidad tal de puntos que permitir√° recuperar la se√±al original, sin p√©rdida de informaci√≥n.

Esto quiere decir que si la se√±al es, por ejemplo, sonido audible por los seres humanos (entre 20 Hz y 20 kHz), para poder representarla de manera que no se pierda informaci√≥n, se necesitar√°n al menos 40 kHz de frecuencia de muestreo. Esto significa que el vector donde se guarde el sonido deber√° tener al menos 40.000 valores num√©ricos por cada segundo de se√±al.

## Ajuste de curvas
Crea un trendline de los datos que tenemos.
por ejemplo, si tenemos un set de datos que se comporta como una funcion lineal, podemos ajustar esos datos y obtener una lineal que modele nuestra data de manera automatica

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy
from scipy import optimize

def function(x, a, b):
   return a*x + b

x = np.linspace(start=-50, stop=10, num=40)
y = function(x, 6, 2)

np.random.seed(62)
noise = 20*np.random.normal(size=y.size)
y = y + noise

popt,cov = scipy.optimize.curve_fit(function, x, y)
a,b= popt

x_new_value = np.arange(min(x), 30, 5)
y_new_value = function(x_new_value, a, b)

plt.scatter(x,y,color="green")
plt.plot(x_new_value,y_new_value,color="red")
plt.xlabel('X')
plt.ylabel('Y')
print("Estimated value of a : "+ str(a))
print("Estimated value of b : " + str(b))
plt.show()
```
