```python
# Ejercicio 2
#ùëÅ' = ùëü. ùëÅ. (1 ‚àí ùëÅ/ùêæ).

# k es el atractor, y r es la "reproduccion"???

n=10
r = 10
k= 300
t=0
n2 = r * n * (1- n / k ) 
dt = 0.01

while t < 3:
  plt.figure(1)
  plt.plot(n,n2,'.r')
  plt.figure(2)
  plt.plot(t,n,'.g')

  n = n + (n2 * dt)
  
  n2 = r * n * (1- n / k ) 

  t = t + dt
```

```python
# diagrama de bifurcaciones
# n(k)


while k<1000:

  n=10
  r= 10
  t=0
  n2 = r * n * (1- n / k ) 
  dt = 0.01

  while t < 3:
    n = n + (n2 * dt)
    
    n2 = r * n * (1- n / k ) 

    t = t + dt
  plt.plot(k,n,'.b')
  k=k+1
```
