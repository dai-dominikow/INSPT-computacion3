```python
import math
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = 270
l = 1
t= 0
g= 9.8
theta = x / l
v=0
a =  0
dt= 0.01

#si varias un poco las condiciones iniciales varia todo el sistema


while t<30:
  plt.figure(1)
  plt.plot(t,x,".")
  plt.figure(2)
  plt.plot(x,v,".")

  #metodo de Cromer: ya que el metodo de Euler no funciona para osciladores, se usa el metodo de Cromer, basicamente se invierte el orden de calculo, primero la V y despues X 
  v = v + a * dt

  x = x + v * dt

  theta = x * l
  
  a =  g * math.sin(theta) 
  
  t=t+dt
```

En vez de usar puntos y graficar uno por uno en loop, es mejor meter toda la data en listas o dataframes y ahi graficar todo prolijamente como una linea continua

```python
l = 1
g= 9.8
theta = x / l
dt= 0.01

tf= 30 

t = np.arange(0 , tf + dt ,dt)
x = np.zeros(len(t))
v = np.zeros(len(t))
a = np.zeros(len(t))

t_len = len(t) - 1

i=0

x[i]= 1.4

while i < t_len :

  #metodo de Cromer: ya que el metodo de Euler no funciona para osciladores, se usa el metodo de Cromer, basicamente se invierte el orden de calculo, primero la V y despues X 
  
  v[i+1] = v[i] + a[i] * dt

  x[i+1] = x[i] + v[i+1] * dt

  theta[i+1] = x[i+1] * l
  
  a[i+1] =  g * math.sin(theta[i+1]) 
    
  i=i+1

plt.figure(1)
plt.plot(t,x)

plt.figure(2)
plt.plot(v,x)
```
