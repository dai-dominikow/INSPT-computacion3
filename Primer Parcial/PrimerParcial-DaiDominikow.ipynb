```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#parámetros a variar
a= 0.2
b= 0.2
c= 15
#color de los ploteos
m= 'teal'
```

```python
#variables estáticas
dt= 0.001
tf= 400
t = np.arange(0 , tf + dt ,dt)
x = np.zeros(len(t))
x1 = np.zeros(len(t))
y = np.zeros(len(t))
y1 = np.zeros(len(t))
z = np.zeros(len(t))
z1 = np.zeros(len(t))
i=0
x[i]=0.1
y[i]=0.1
z[i]=0.1
x1[i]=0
y1[i]=0
z1[i]=0
t_len = len(t) - 1
```

```python
while i < t_len:

  x1[i+1] = - y[i] - z[i]
  y1[i+1] = x[i]  + (a * y[i])
  z1[i+1] = b + z[i] * (x[i] - c)
  x[i+1] = x[i] + x1[i+1] * dt
  y[i+1] = y[i] + y1[i+1] * dt
  z[i+1] = z[i] + z1[i+1] * dt
  i = i+1
```

```python
figure, axis = plt.subplots(1,3,figsize=(15,5))
axis[0].plot(x, y,color = m)
axis[0].set_title("plano x y - " + "a: " + str(a)  + ", b: " + str(b) +", c: " + str(c))
axis[1].plot(y, z,color = m)
axis[1].set_title("plano y z - " + "a: " + str(a)  + ", b: " + str(b) +", c: " + str(c))
axis[2].plot(x, z,color = m)
axis[2].set_title("plano x z - " + "a: " + str(a)  + ", b: " + str(b) +", c: " + str(c))
plt.show()
```

```python
figure, axis = plt.subplots(1,3,figsize=(15,5))
axis[0].plot(x1, y1, color = m)
axis[0].set_title("plano x' y' - " + "a: " + str(a)  + ", b: " + str(b) +", c: " + str(c))
axis[1].plot(y1, z1, color = m )
axis[1].set_title("plano y' z' - " + "a: " + str(a)  + ", b: " + str(b) +", c: " + str(c))
axis[2].plot(x1, z1, color = m )
axis[2].set_title("plano x' z' - " + "a: " + str(a)  + ", b: " + str(b) +", c: " + str(c))
plt.show()
```

```python
figure, axis = plt.subplots(1,3,figsize=(15,5))
axis[0].plot(t, x,color = m)
axis[0].set_title("evolucion t x - " + "a: " + str(a)  + ", b: " + str(b) +", c: " + str(c))
axis[1].plot(t, y,color = m)
axis[1].set_title("evolucion t y - " + "a: " + str(a)  + ", b: " + str(b) +", c: " + str(c))
axis[2].plot(t, z,color = m)
axis[2].set_title("evolucion t z - " + "a: " + str(a)  + ", b: " + str(b) +", c: " + str(c))
plt.show()
```

```python
figure, axis = plt.subplots(1,3,figsize=(15,5))
axis[0].plot(t, x1,color = m)
axis[0].set_title("evolucion t x' - " + "a: " + str(a)  + ", b: " + str(b) +", c: " + str(c))
axis[1].plot(t, y1,color = m)
axis[1].set_title("evolucion t y' - " + "a: " + str(a)  + ", b: " + str(b) +", c: " + str(c))
axis[2].plot(t, z1,color = m)
axis[2].set_title("evolucion t z' - " + "a: " + str(a)  + ", b: " + str(b) +", c: " + str(c))
plt.show()
```

```python
figure, axis = plt.subplots(1,3,figsize=(15,5))
axis[0].plot(x1, x,color = m)
axis[0].set_title("Diag.  de fase  x' x  - " + "a: " + str(a)  + ", b: " + str(b) +", c: " + str(c))
axis[1].plot(y1, y,color = m)
axis[1].set_title("Diag. de fase  y' y  - " + "a: " + str(a)  + ", b: " + str(b) +", c: " + str(c))
axis[2].plot(z1, z,color = m)
axis[2].set_title("Diag.  de fase  z' z  - " + "a: " + str(a)  + ", b: " + str(b) +", c: " + str(c))
plt.show()
```
