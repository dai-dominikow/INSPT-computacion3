```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
# dependencia de las condiciones iniciales 
# divergencia de trayectorias

#el caos eiste siempre que n >= 3 no lineales 

# Declaro variables

a= 10 
b= 28
c= 2.666

dt= 0.01
tf= 40

t = np.arange(0 , tf + dt ,dt)
x = np.zeros(len(t))
x1 = np.zeros(len(t))
y = np.zeros(len(t))
y1 = np.zeros(len(t))
z = np.zeros(len(t))
z1 = np.zeros(len(t))

i=0

x[i]=1.0
y[i]=1.0
z[i]=1.0
x1[i]=0
y1[i]=0
z1[i]=0

t_len = len(t) - 1
```

```python
i=0
while i < t_len:

  x1[i+1] = a * (y[i] - x[i])
  y1[i+1] = x[i]  * (b - z[i]) - y[i]
  z1[i+1] = (x[i] * y[i]) - (c * z[i] )

  x[i+1] = x[i] + x1[i+1] * dt
  y[i+1] = y[i] + y1[i+1] * dt
  z[i+1] = z[i] + z1[i+1] * dt



  i = i+1
```

```python
plt.figure(1)
plt.plot(t,x)
plt.figure(2)
plt.plot(t,y)
plt.figure(3)
plt.plot(t,z)
```
