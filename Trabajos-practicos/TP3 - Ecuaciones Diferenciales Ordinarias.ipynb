```python
import numpy as np
import math as math
import matplotlib.pyplot as plt
import pandas as pd
```

```python
n = 2.66e21
m = 1
l= -1.38e-11 * 3600* 24 * 365 #como la unidad es 1/s, y son 1600 años, vamos a pasar todo esto a años. 
t=0
dt = 1 # el paso en años
n_medio=n/2

#dn/dt = -l * n  

while n > n_medio:
  # aproximo dn/dt = l * n
  plt.plot(t,n,'.')  
  n = n + ( l * n * dt)
  t=t+dt
print(t)  
```

```python
 #  ejercicio 2  si quiero un tiro vertical:  es una edo de grado dos, tengo que pasar la derivada de grado dos de x, a la derivada de v 
x=0 
v= 15 # dx/dt
t= 0
t_max =5
g=-9.8 # d2x/dt2
dt=0.1


while t < t_max:
  plt.plot(t,x,'.r')
  plt.plot(t,v,'.b')
  x = x + (v * dt)
  v = v + (g * dt)
  t = t + dt
print(v)
```
