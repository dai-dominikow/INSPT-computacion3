### 1) Matemática:<br> 
Hallar la porción $d$ de una esfera de radio r=10cm y densidad ρ=0,638g/cm3​ ​ que queda sumergida en agua de manera estable. El modelo matemático es<br>
$\frac{\pi(d^{3}​ ​- 3d^{2}​ r + 4r^{3}​ ​\rho )}{3} = 0$


```python
def f(d):
  r = 0.10
  rho = 638
  y = (pi*(d**3 - 3*r*(d**2)+ 4*(r**3)*rho))/3
  return y 
```

```python
a = -2
b = 3
d = np.linspace(a,b,20)
plt.plot(d,f(d))
plt.grid()
plt.show()
```

```python
error= 10
i=0
while error >1e-8:
  c= (a+b)/2
  fa= f(a)
  fc= f(c)
  if fc==0:
    raiz = c
    break
  elif fa*fc <0 :
    b = c
  else:
    a=c
  raiz=c
  i+=1
  error = abs(fc)

print("la raiz (porcentaje sumergido) es:", round(raiz,5))
print("salió en ",i,"iteraciones")
```

### 2. Gases:<br>
La ecuación de Van Der Waals para gases reales a grandes presiones es<br>
$(p + \frac{a n^{2}}{V^{2}})(V - n b) = n R T$<br>
Donde $a$ y $b$ son constantes empíricas propias de cada gas.<br>
Calcular el volumen del gas si se tiene un mol de nitrógeno gaseoso en un tanque a 0ºC y 400 atm de presión, siendo $a = 1,3896 \frac{atm l^{2}}{mol^{2}}$, $b = 0,03913 \frac{l}{mol}$ y $R = 0,082 \frac{atm l}{mol K}$


```python
def f(V):
    a = 1.3896
    b = 0.03913
    R = 0.082
    p = 400
    T = 273
    n = 1
    V = ((p + (a * (n**2) )/ V**2 )*( V - n * b) ) - (n * R * T)
    return V
```

```python
a = 0.03
b = 0.08
V = np.linspace(a,b,40)
plt.plot(V,f(V))
plt.grid()
plt.show()
```

```python
error= 10
i=0
while error >1e-8:
  c= (a+b)/2
  fa= f(a)
  fc= f(c)
  if fc==0:
    raiz = c
    break
  elif fa*fc <0 :
    b = c
  else:
    a=c
  raiz=c
  i+=1
  error = abs(fc)

print("la raiz (Volumen del gas) es:", round(raiz,5))
print("salió en ",i,"iteraciones")
```

### 3. Cinemática elegante: <br>

Un tiro vertical a baja velocidad considerando la viscosidad del aire puede modelarse con la ecuación horaria (considerando $C=\frac{m}{\beta}$ siendo $\beta$ el coeficiente que multiplica la velocidad en la ecuación $F_v=\beta v$ para la fuerza de arrastre). Siendo C=2 s, la altura inicial 10 m y la velocidad de lanzamiento 18 m/s, determinar el tiempo de vuelo. <br>
$y = y_0 + (C v_y + C^{2} g).(1 − e^{− \frac{t}{C}}) − g C t$<br>
donde<br>
$g = 9,8\frac{m}{s^{2}}$

```python
#f_v es proporcional a la velocidad
def f2(t):
  C= 2 
  h=10
  v0=18
  g=9.8
  y= h +(C * v0 + C ** 2 * g)* (1 - math.e ** (-t/C)) - g * C * t
  return y
a = 0
b = 5
t= np.linspace(a,b,50)
plt.plot(t,f2(t))
plt.grid()
plt.show()
```

```python
error= 10
i=0
while error >1e-8:
  c= (a+b)/2
  fa= f2(a)
  fc= f2(c)
  if fc==0:
    raiz = c
    break
  elif fa*fc <0 :
    b = c
  else:
    a=c
  raiz=c
  i+=1
  error = abs(fc)

print("la raiz (tiempo de vuelo) es:", round(raiz,5))
print("salió en ",i,"iteraciones")
```
