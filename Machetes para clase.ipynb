Derivadas

```python
def deriv(x,t):

  j= len(x) - 1
  i=0
  dx=np.zeros(len(x))
  
  while i < j:
    dx[i]= (x[i+1]-x[i])/(t[i+1]-t[i])
    i=i+1
  dx[i]=None

  return dx
  
v = deriv(x,t)
plt.plot(t, v, color= 'green')
plt.xlabel('tiempo')
plt.ylabel('velocidad')
plt.show()
```

# Simulación dinámica

Los sistemas dinámicos o el estudio dinámico de los sistemas es el que refiere a sistemas que evolucionan en el tiempo.
Retrato de fase: ejemplo velocidad vs posición, la derivada de una variable respecto a dicha variable. Representacion geometrica de las trayectorias de un sistema dinamico en el plano.
Los retratos de fases revelan informacion sobre la existencia de atractores y repulsores.
Los sistemas dinámicos no lineales exhiben atractores, que se definen como puntos en el retrato de fases donde todas las trayectorias cercanas convergen, donde tiende a evolucionar el sistema.

### metodo de Euler:
Metodo de Euler:  CONSISTE EN SEGUIR LAS TANGENTES DE LA FUNCION EN CADA PUNTO se utiliza cuando tenemos la derivada de la funcion que representa la solucion de una EDO, es decir, su pendiente en cualquier punto. Este método es un método de primer orden, lo que quiere decir que cada variable de estado se calculará a partir de su valor en el tiempo anterior.
Un lector entrenado en matemática, notará que estamos calculando una magnitud a partir de utilizar la derivada de la magnitud (v=dy/dt) con respecto al tiempo, multiplicada por el incremento del tiempo. Es decir, uniendo el punto conocido con el punto a calcular a partir de
una recta que tiene como pendiente la derivada de la posición en ese punto.

Metodo de integracion numerica para resolver ecuaciones diferenciales ordinarias a partir de un valor inicial dado.
Cada instante calculo la derivada de la posiciòn con la derivada del punto anterior.
Sirve solamente para ecuaciones diferenciales de grado uno.
posicion --> velocidad --> aceleracion
Es un metodo de primer orden, lo que significa que el error local (diferencia que se produce entre el valor real de la función y el aproximado mediante la recta tangente, en lugar de moverse por la curva, suponiendo que el punto donde se cruzan la curva real y la recta que la aproxima no tiene error alguno) es proporcional al cuadrado del paso y el erorr global es proporcional al tamaño del paso (en general el error global tiene un orden menos que el local, es un poco mas grueso ya que los errores locales se acumulan dand lugar al global).

### metodo de Cromer:
La diferencia con el método de Euler estándar es que el método de Euler semi-implícito usa v_n+1 en la ecuación para x_n+1 , mientras que el método de Euler usa v_n.
Es mas estable en problemas de orbitas y oscilatorios, pero en cuanto a precision es igual de pobre.
Se calcula en otro orden
aceleracion --> velocidad --> posicion

### FFT
La transformada rápida de Fourier, conocida por la abreviatura FFT (del inglés Fast Fourier Transform) es un algoritmo eficiente que permite calcular la transformada de Fourier discreta (DFT) y su inversa. La FFT es de gran importancia en una amplia variedad de aplicaciones, desde el tratamiento digital de señales y filtrado digital en general a la resolución de ecuaciones en derivadas parciales o los algoritmos de multiplicación rápida de grandes enteros. 
Esta transformada convierte la señal del dominio del tiempo al dominio de la frecuencia,  devolviendo el espectro (Es decir, armónicos de qué frecuencia componen la señal).
la transformada de Fourier contiene todas las frecuencias del tiempo durante el cual existió la señal; es decir, en la transformada de Fourier se obtiene un solo espectro de frecuencias para toda la función.
La antitransformada de Fourier permite, teniendo el espectro, recuperar la función original. Esto es muy útil porque permite realizar operaciones sobre el espectro que serían más difíciles de realizar en el dominio del tiempo. En el dominio de la frecuencia se pueden aplicar filtros para recortar o “podar” (es el término técnico) toda una frecuencia del sonido. El resultado de la transformada de Fourier es una función compleja. 

Dada una señal periodica, podemos usar la transformada rapida de fourier para ver su contenido de frecuencias.
Graficamente, pasamos de una funcion dependiente del tiempo a una dependiente de las frecuencias en el plano complejo. El plot que hacemos solo refleja la parte real del complejo aunque tambien se puede laburar con la imaginaria, pero vamos a usar RFFT (que devuelve la parte real)
La transformada entonces reemplaza integrales por sumatorias (similar al metodo de los trapecios)

Un muestreo puede ser representado como una multiplicacion de la señal original (continua) con un pulso unitario.

¿Qué quiere decir muestreada? que la señal real por ser analógica tendría infinitos valores, pero nosotros no podemos representar infinitos valores en la pc, por lo que la frecuencia de muestreo establece cuántos valores de tensión, intensidad o lo que sea que represente la señal se tomarán por segundo.

La frecuencia de muestreo debe cumplir con el teorema del Muestreo, o teorema de Nyquist-Shannon que explica la relación entre la velocidad de muestreo y la frecuencia de la señal medida.
Básicamente dice que para reconstruir una señal, el muestreo debe tener el doble de frecuencia que la máxima de la señal.

Si una señal es muestreada a una velocidad de muestreo menor que el doble de la frecuencia Nyquist, los componentes de frecuencias falsas más bajas aparecen en los datos muestreados. Este fenómeno se conoce como aliasing. Se produce cuando una velocidad de muestreo es demasiado baja y reproduce una representación incorrecta de forma de onda.

Este teorema es muy específico y habla de que si una señal periódica analógica tiene una banda de frecuencias definida, entonces muestreando con una frecuencia que es el doble de la frecuencia máxima de la señal se obtendrá una cantidad tal de puntos que permitirá recuperar la señal original, sin pérdida de información.

Esto quiere decir que si la señal es, por ejemplo, sonido audible por los seres humanos (entre 20 Hz y 20 kHz), para poder representarla de manera que no se pierda información, se necesitarán al menos 40 kHz de frecuencia de muestreo. Esto significa que el vector donde se guarde el sonido deberá tener al menos 40.000 valores numéricos por cada segundo de señal.

## Ajuste de curvas
Crea un trendline de los datos que tenemos.
por ejemplo, si tenemos un set de datos que se comporta como una funcion lineal, podemos ajustar esos datos y obtener una lineal que modele nuestra data de manera automatica

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy
from scipy import optimize

def function(x, a, b):
   return a*x + b

x = np.linspace(start=-50, stop=10, num=40)
y = function(x, 6, 2)

np.random.seed(62)
noise = 20*np.random.normal(size=y.size)
y = y + noise

popt,cov = scipy.optimize.curve_fit(function, x, y)
a,b= popt

x_new_value = np.arange(min(x), 30, 5)
y_new_value = function(x_new_value, a, b)

plt.scatter(x,y,color="green")
plt.plot(x_new_value,y_new_value,color="red")
plt.xlabel('X')
plt.ylabel('Y')
print("Estimated value of a : "+ str(a))
print("Estimated value of b : " + str(b))
plt.show()
```
